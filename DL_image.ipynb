پکیج های مورد **نیاز**

```python
import numpy
from tensorflow import keras
from keras.constraints import maxnorm
from sklearn.preprocessing import OneHotEncoder
from keras.models import Sequential, load_model
from keras.layers import Conv2D, MaxPool2D, Dense, Dropout, Flatten
from keras.layers import BatchNormalization
from keras.callbacks import EarlyStopping
from sklearn.metrics import confusion_matrix
from keras.utils import np_utils
from keras.datasets import cifar10
import matplotlib.pyplot as plt
```

لود کردن دیتاست و اختصاص دیتا به بخش آموزش و تست

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_test.shape
```

```python
y_train[0]
```

--------------------------------------------

تصویر نمونه 32*32 پیکسل

```python
plt.imshow(X_train[0])
```

--------------------------

تغییر مقیاس بین صفر و **یک**

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train = X_train / 255.0
X_test = X_test / 255.0
```

وان هات انکدینگ برای جلوگیری از ارتباط بین کلاس **ها**

```python
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

```python
y_train[0]
```

```python
num_class = y_test.shape[1] #10
```

```python
X_train.shape[1:]
```

کاهش حجم و انتخاب دیتا با اهمیت با کانولوشن و مکس پولینگ
جلوگیری از اورفیتینگ با دراپ اوت و بچ نرمالزیشن
ایجاد لایه پنهان با اکتیویشن رلو و لایه خروجی با اکتیویشن سافت مکس

```python
model = Sequential()
model.add(Conv2D(32, (3,3), padding='same',
                 input_shape=X_train.shape[1:], activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Conv2D(64, (3,3), padding='same', activation='relu'))
model.add(MaxPool2D(2))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Conv2D(64, (3,3), padding='same', activation='relu'))
model.add(MaxPool2D(2))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Conv2D(128, (3,3), padding='same', activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Flatten())
model.add(Dropout(0.2))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(num_class, activation = 'softmax'))
```

کامپایل کردن ممدل

```python
model.compile(loss='categorical_crossentropy',
              optimizer = 'adam', metrics = ['accuracy'])
```

نمای کلی از لایه ها

```python
model.summary()
```

آموزش مدل با سی دور تکرار و بسته های آموزش 64 تایی

```python
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=30, batch_size=64)
```

```python
for key,val in history.history.items():
  print(key)
```

نمایش دقت مدل به وسیله ی ماتریس اختشاش

```python
predictions = model.predict(X_test)

predictions = np.argmax(predictions, axis = 1)
rounded_labels=np.argmax(y_test, axis=1)

cm = confusion_matrix(rounded_labels, predictions)

plt.figure(figsize=(9,9))
sns.heatmap(cm, cbar=False, xticklabels=labels, yticklabels=labels,
            fmt='d', annot=True, cmap=plt.cm.Blues)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
